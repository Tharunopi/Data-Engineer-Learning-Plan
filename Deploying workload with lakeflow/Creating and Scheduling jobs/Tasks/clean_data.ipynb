```python
import pyspark.sql.functions as F
```

```python
%sql
USE CATALOG dbacademy;
USE SCHEMA controlling_job_flow;
```

```python
columns = ["year", "version", "statefips", "countyfips", "geocat", "agecat", "racecat", "sexcat"]
```

```python
df1 = (spark.read.table("input_1")).select(columns)
df2 = (spark.read.table("input_2")).select(columns)
df3 = (spark.read.table("input_3")).select(columns)
```

```python
df1.show()
df2.show()
df3.show()
```

```python
df1 = df1.withColumn("version", (F.col("year") + 0.1).cast("string"))
df2 = df2.withColumn("version", (F.col("year") + 0.1).cast("string"))
df3 = df3.withColumn("version", (F.col("year") + 0.1).cast("string"))
```

```python
df1.fillna(-1)
df2.fillna(-1)
df3.fillna(-1)
```

```python
display(df1)
```

```python
df1.write.mode("overwrite").format("delta").saveAsTable("dbacademy.controlling_job_flow.input_1")
df2.write.mode("overwrite").format("delta").saveAsTable("dbacademy.controlling_job_flow.input_2")
df3.write.mode("overwrite").format("delta").saveAsTable("dbacademy.controlling_job_flow.input_3")
```
