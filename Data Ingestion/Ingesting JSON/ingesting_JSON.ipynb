# **Creating a bronze level table**

```python
%sql
DROP TABLE IF EXISTS json_bronze;

CREATE TABLE json_bronze AS 
SELECT * FROM read_files(
  '/Volumes/dbacademy/first_data/json_volume/converted-file.json',
  format => 'text'
);

SELECT * FROM json_bronze
LIMIT 10
```

# **Flattening JSON string columns**

```python
%sql
SELECT value:Name, value:Sex FROM json_bronze
```

# **Getting Schema of JSON**

```python
%sql
SELECT schema_of_json('{"PassengerId":1,"Survived":0,"Pclass":3,"Name":"Braund, Mr. Owen Harris","Sex":"male","Age":22,"SibSp":1,"Parch":0,"Ticket":"A\/5 21171","Fare":7.25,"Cabin":null,"Embarked":"S"}') AS schema
```

```python
%sql
DROP TABLE IF EXISTS json_struct_bronze;

CREATE TABLE json_struct_bronze AS 
SELECT from_json(value, 'STRUCT<Age: BIGINT, Cabin: STRING, Embarked: STRING, Fare: DOUBLE, Name: STRING, Parch: BIGINT, PassengerId: BIGINT, Pclass: BIGINT, Sex: STRING, SibSp: BIGINT, Survived: BIGINT, Ticket: STRING>') AS value
FROM json_bronze;

SELECT * FROM json_struct_bronze LIMIT 5
```

```python
%sql
SELECT value.Age, value.Name FROM json_struct_bronze
ORDER BY value.Age
```

# **Explode Arrays**

```python
%sql
SELECT EXPLODE_OUTER(value.items) FROM json_struct_bronze
```

# **Working with variant columns**

```python
%sql
SELECT parse_json(value) AS value FROM json_bronze
```
